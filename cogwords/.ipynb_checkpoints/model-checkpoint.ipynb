```python
class Encoder:
    def __init__(self, dim_input, dim_output):
        self.dense = dense(dim_input, dim_output, weight_initializer = glorot_normal, bias = False)
        
    def __call__(self, x):
        return self.dense(x)
    
    @property    
    def parameters(self):
        return self.dense.parameters
```

```python
train_images = [*resnet18_features.keys()][0:int(len(resnet18_features.keys()) * 0.8)]
test_images = [*resnet18_features.keys()][-int(len(resnet18_features.keys()) * 0.2) :]
```

```python
def get_confusor(image_id):
    confusor_id = np.random.choice(train_images)
    while confusor_id == image_id:
        confusor_id = np.random.choice(train_images)
    return confusor_id
```

```python
train_triples = []
for image in train_images:
    for caption in db.get_caption_ID(image):
        caption_id = caption
        confusor_id = get_confusor(image)
        train_triples.append((caption_id, image, confusor_id))
```
